```python
import numpy as np
import matplotlib.pyplot as plt
plt.ylabel('Likehood ratio')
plt.xlabel("X")
x=np.arange(-10,10,0.01)
x1=[0.5,0.5]
y1=0.5*np.exp(-1*abs(x))
y2=0.5*np.exp(-1*abs(x-1))
y3=np.exp(abs(x-1)-abs(x))
plt.plot(x,y3,label="Likehood ratio")
plt.plot(x1,[0,1],label="x=1/2 Decision boundary")
plt.plot(x,y1,label="P(x|w1)")
plt.plot(x,y2,label="P(x|w2)")
plt.legend()
plt.show()
```
